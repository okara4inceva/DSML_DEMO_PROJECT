```python
import pandas as pd

df = pd.read_csv("../data/heart_failure_clinical_records_dataset.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.violinplot(x="DEATH_EVENT", y="ejection_fraction", data=df)

plt.title("Ejection Fraction distribution by mortality")
plt.show()
```

```python
sns.violinplot(x="DEATH_EVENT", y="ejection_fraction", data=df, inner="quartile")
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True), 
            annot=True, 
            cmap="coolwarm",
            fmt=".2f")

plt.title("Feature Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x="DEATH_EVENT", y="age", data=df)

plt.title("Age distribution by mortality")
plt.xlabel("Death Event")
plt.ylabel("Age")

plt.show()
```

```python
sns.violinplot(x="DEATH_EVENT", y="serum_creatinine", data=df)

plt.title("Serum Creatinine by mortality")

plt.show()
```

### The violin plot shows that patients who died tend to have higher serum creatinine levels compared to survivors. This suggests that impaired kidney function is associated with increased mortality risk in heart failure patients.
